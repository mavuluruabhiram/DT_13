## Sort Multi_Index DataFrame

```python
neighborhoods.sort_index(level = 2) # Based On Street
```

```python
neighborhoods.sort_index(level = [1,2]) # Based On City and state
```

```python
# For City Ascending , For Street Descemding

neighborhoods.sort_index(level = [1,2] , ascending=[True , False]).head()
```

```python
# Sorted based on Columns : category descending  - Service and Culture  

neighborhoods.sort_index(axis=1 , level='Category' , ascending=False).head() 
```

```python
neighborhoods.sort_index(axis=1 , level='Category' , ascending=True).head() 
```

```python
neighborhoods.sort_index(axis=1 , level='Subcategory' , ascending=True).head() 
```

```python
neighborhoods.sort_index(axis=1 , level='Subcategory' , ascending=False).head() 
```

## CrossSection

```python
# Selecting MO State ( ROWS )

neighborhoods.xs(level = 'State' , key = 'MO')
```

```python
# COLUMNS

neighborhoods.xs(axis = 1 , level = 'Subcategory' , key = 'Schools')
```
