### Replacing - converting Object to FLoat by using Replace()

- Some characters (like $, ., ^, *) have special meanings in regex.

- In regex:

    - $ means “end of string”, not the dollar sign.

    - , is fine, but others like . or * can behave unexpectedly.


In pandas, a column like movies['Gross'] is a Series.

If that Series contains strings (like "$1,234,567"), you cannot directly use Python string methods (like .replace()) on the whole Series.

Without .str, you’d get an AttributeError, because a Series doesn’t have .replace in the same sense as Python strings.



```python
movies['Gross'] = (
    
movies['Gross']
.str.replace('$' , '' , regex=False)
.str.replace(',' , '' , regex=False)    


).astype(float)

# regex : regular expression - used to search data and match columns
```

### Filtering -  dropna() Method --------> used to drop missing values

- dropna(how = 'any') -------- > drop the rows which has atleast one NaN (missing) Value
- dropna(how = 'all') -------- > drop the rows which has all NaN (missing) Values

- dropna(subset = ['COLUMN NAME ']) ------------- > Drop row which has missing values in particular column 
    - dropna(subset = ['Team'])

- employees.dropna(thresh = NUM) ---------- > Drops Row which Has 'No.of NON - MISSING  values 
    - employees.dropna(thresh = 1) -----> 1 NON-MISSING VALUES

```python
employees.head()
```

```python
employees.dropna().head()
```

```python
#SAME

employees.dropna(how = 'any').head()
```

```python
employees.dropna(how = 'all').head()
```

```python
employees.dropna(subset = ['Team'])
```

```python
employees.dropna(thresh = 1) # 1 NON-MISSING VALUES

employees.dropna(thresh = 2) # 2 NON-MISSING VALUES
```

### Filtering - duplicated() Method 
- TRUE for  repeated value
- FALSE for NOT repeated

```python
employees['Team'].head(10)
```

```python
employees['Team'].head(10).duplicated()
```

```python
# INVERSION 

~employees['Team'].head(10).duplicated()
```

```python
# DATA OF REPEATED TEAMS

employees[employees['Team'].duplicated()]

```

```python
# DATA OF NON - REPEATED TEAMS

employees[~employees['Team'].duplicated()]

```

```python
# Displays Non - Repeating Values In Last

employees[employees['Team'].duplicated(keep='last')]
```

```python
# Displays Repeating Values In First

employees[employees['Team'].duplicated(keep='first')]
```
