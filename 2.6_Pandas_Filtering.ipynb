### Replacing - converting Object to FLoat by using Replace()

- Some characters (like $, ., ^, *) have special meanings in regex.

- In regex:

    - $ means “end of string”, not the dollar sign.

    - , is fine, but others like . or * can behave unexpectedly.


In pandas, a column like movies['Gross'] is a Series.

If that Series contains strings (like "$1,234,567"), you cannot directly use Python string methods (like .replace()) on the whole Series.

Without .str, you’d get an AttributeError, because a Series doesn’t have .replace in the same sense as Python strings.



```python
movies['Gross'] = (
    
movies['Gross']
.str.replace('$' , '' , regex=False)
.str.replace(',' , '' , regex=False)    


).astype(float)

# regex : regular expression - used to search data and match columns
```

### Filtering - Apply func

```python
pokemon = pd.read_csv('pokemon.csv')
```

```python
pokemon.head()
```

```python
pokemon = pd.read_csv('pokemon.csv' , index_col = 'Pokemon').squeeze()

pokemon
```

```python
def single_or_multi(pokemon_type):
    if "/" in pokemon_type:
        return "Multi"
    
    return "Single"

pokemon.apply(single_or_multi)
```
